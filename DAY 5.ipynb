```python
y1 = np.log(1 + 2*x)
y2 = x ** 2
# y3 = x
# plt.plot(x, y3, label = 'straightLine', c = 'green')
plt.scatter(x, y1, label = 'scatter1', c = 'brown')
plt.scatter(x, y + 1, label = 'scatter2')
plt.plot(x, y1, label = 'log', c = 'gold')
plt.plot(x, y2, label = 'power', c = 'black', ls = '--', lw = 2, marker = 's', mfc = 'r')
plt.scatter(x, y2 + 1, label = 'scatter3')
plt.title('Some Functions')
plt.xlabel('x')
plt.ylabel = ('f(x)')
plt.legend()
plt.show()
```
